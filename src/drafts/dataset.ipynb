```python
new_trade_data = pd.read_clipboard()

new_trade_data
```

```python
new_trade_data["date"] = pd.to_datetime(new_trade_data["date"])

new_trade_data.to_csv("trade-data.csv", index=False)
```

```python
melt_new_trade_data = new_trade_data.melt(["date"])
melt_new_trade_data["type"] = [ticker[1:3] for ticker in melt_new_trade_data["variable"]]
melt_new_trade_data["country_code"] = [ticker[4:].split(" ")[0] for ticker in melt_new_trade_data["variable"]]

melt_new_trade_data
```

```python
melt_new_trade_data["type"].unique()
```

```python
relative_trade_data = []
for country in melt_new_trade_data["country_code"].unique():
    country_data = melt_new_trade_data.loc[melt_new_trade_data["country_code"] == country]
    country_trade_data = country_data.pivot_table(index=["date"], columns=["type"], values=["value"])
    country_trade_data.columns = country_trade_data.columns.droplevel()

    tmp_relative_trade_data = pd.DataFrame(country_trade_data["IM"] / country_trade_data["EM"],
                                           columns=[country])

    relative_trade_data.append(tmp_relative_trade_data)
relative_trade_data = pd.concat(relative_trade_data, axis=1)
relative_trade_data = relative_trade_data.reset_index()
```

```python
relative_trade_data.to_csv("relative-trade-data.csv", index=False)
```
